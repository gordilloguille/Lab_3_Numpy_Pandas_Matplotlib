## **2) Realice una gráfica de barras sobre la población de cada ciudad, ordenela de menor a mayor y responda:**

a. ¿Cuál es la segunda ciudad más poblada?


```python
# Ordenamos la población de menor a mayor
poblacion = dataset.sort_values("Poblacion", ascending=True)
# Reduje el tamaño de registros ya quue dejaba de ser discreta ya que los municipios sobrepasaban los 30, ya con una grafica de barras se miraba mal
# al reducir los top 30 municipios en base a su poblacion es mas evidente quienes son los que tienen mas poblacion
poblacion = poblacion[poblacion["Poblacion"]>=poblacion["Poblacion"].iloc[-30]]
poblacion = poblacion.sort_values("Poblacion")

# Creamos la gráfica de barras
plt.bar(poblacion["Municipio"], poblacion["Poblacion"])
plt.title("población de cada Municipio TOP 30")
plt.xlabel("Ciudad")
plt.ylabel("Población")
plt.xticks(rotation=90)
plt.show()

# Obtenemos la segunda ciudad más poblada
segunda_ciudad = poblacion["Municipio"].iloc[-2]
# Imprimimos la segunda ciudad más poblada
print(f"La segunda ciudad más poblada es: {segunda_ciudad}")
```

## **8) Determine los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.**

```python
#Instalo el packete de geopy para calcular las distancias
!pip install geopy
```

```python
import geopy
from geopy.distance import geodesic

# Obtener el indice del monicipio de Madrid del frame para obtener los valores de "X" y "Y"
municipio_madrid = dataset[(dataset["Provincia"] == "Madrid") &  (dataset["Municipio"]== "Madrid")].index[0]

#con el atributo loc obtenemos el valor de las celdas donde estan las coordenadas, como el indice ya encontrado
madrid_X = dataset.loc[municipio_madrid,"X"]
madrid_Y = dataset.loc[municipio_madrid,"Y"]

# Creo una columna para almacenar las distancias y la lleno con valores nulos
# utilizando np.nan para lo cual es un valor especial de la librería NumPy que representa el valor nulo
dataset["Distancia_Madrid"] = np.nan

for i in range(len(dataset)):
    municipio_lat = dataset.loc[i, 'Y']
    municipio_lon = dataset.loc[i, 'X']
    
    # Calculao la distancia usando la fórmula de Haversine y lo guardo en km
    distancia = geodesic((madrid_Y, madrid_X), (municipio_lat, municipio_lon)).km
    
    # Almaceno la distancia en la columna 'Distancia_Madrid'
    dataset.loc[i, 'Distancia_Madrid'] = distancia
 
# Creo una df quitando a Madrid para que no sea considerado en el analisis
municipios_distancias = dataset[dataset['Municipio'] != "Madrid"]
# Lo ordeno solo para verificaciones posteriores
municipios_distancias = municipios_distancias.sort_values("Distancia_Madrid", ascending=True)
 
# Obtener el municipio con la distancia mínima
municipio_mas_cercano = municipios_distancias.loc[municipios_distancias['Distancia_Madrid'] ==municipios_distancias['Distancia_Madrid'].min(), 'Municipio'].values[0]
municipio_mas_cercano_Km = municipios_distancias.loc[municipios_distancias['Distancia_Madrid'] ==municipios_distancias['Distancia_Madrid'].min(), 'Distancia_Madrid'].values[0]

# Obtener el municipio con la distancia máxima
municipio_mas_lejano = municipios_distancias.loc[municipios_distancias['Distancia_Madrid'] == municipios_distancias['Distancia_Madrid'].max(), 'Municipio'].values[0]
municipio_mas_lejano_Km = municipios_distancias.loc[municipios_distancias['Distancia_Madrid'] == municipios_distancias['Distancia_Madrid'].max(), 'Distancia_Madrid'].values[0]

print(f'El Municipio mas cercano a Madrid es: "{municipio_mas_cercano}" con una distancia de {round(municipio_mas_cercano_Km,4)} kilometros a Madrid.')
print(f'El Municipio mas Lejano a Madrid es: "{municipio_mas_lejano}" con una distancia de {round(municipio_mas_lejano_Km,4)} kilometros a Madrid.')
```
